# Interview Questions Scraper

```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
import re
import requests
from bs4 import BeautifulSoup
import time
import nltk

from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

from IPython.display import HTML

# Configure pandas dataframes to display entire content in column
pd.set_option('display.max_colwidth', 0)
```

```python
def clean_data(df):
    # Clean of everything but aplhabets and whitespaces
    df.questions = df.questions.apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
    # Strip leading whitespaces and covert to lowercase
    df.questions = df.questions.apply(lambda x: x.strip().lower())
    # Tokenize
    df.questions = df.questions.apply(lambda x: word_tokenize(x))
    # Remove stop words
    stop_words = stopwords.words('english')
    df.questions = df.questions.apply(lambda x: [word for word in x if word not in stop_words])
    
```

```python
def parse_html(URL):
    # get the webpage
    r = requests.get(URL)
    # make a soup out of the webpage
    soup = BeautifulSoup(r.text, 'lxml')
    #print (soup_1.prettify())
    return soup

```

```python
URL_1 = 'https://www.dezyre.com/article/100-data-science-interview-questions-and-answers-general-for-2018/184'
soup_1 = parse_html(URL_1)

results = soup_1.find_all('strong')

df_1 = pd.DataFrame(results)
df_1.columns = ['questions']
```

```python
df_1.drop(df_1.index[:2], inplace=True)
df_1.drop(df_1.index[1:6], inplace=True)

#df_1.dropna()
#df_1.questions = df_1.questions.apply(lambda x: re.sub(r'<strong>', '', x) if pd.notnull(x))#0 if x.dtype == NoneType )

df_1.questions = df_1.questions.apply(lambda x: re.sub(r'<strong>', '', x.text))

df_1.questions = df_1.questions.apply(lambda x: x if re.search(r'\d+\)', x) else "blah")

df_1 = df_1[~df_1.questions.str.contains("blah")]

df_1.head
```

```python
clean_data(df_1)

df_1.head
```
